```python
import matplotlib.pyplot as plt

rounds = [1, 2, 3, 4, 5]
commissions = [0.15, 0.18, 0.20, 0.22, 0.24]
investments = [25, 30, 28, 26, 24]
profit_m = [38600, 43440, 45680, 47840, 50064]
profit_s = [88352, 88500, 88265, 87310, 86290]

fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:red'
ax1.set_xlabel('Round')
ax1.set_ylabel('Commission Rate (M)', color=color)
ax1.plot(rounds, commissions, color=color, marker='o', label='Commission Rate', linewidth=2)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Innovation Investment (S)', color=color)
ax2.plot(rounds, investments, color=color, marker='s', linestyle='--', label='Innovation Inv.', linewidth=2)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Platform Game Dynamics: Commission vs. Innovation')
fig.tight_layout()
plt.grid(True, alpha=0.3)
plt.show()
```
